```python
import json
from keybert import KeyBERT
```

```python

# Load the JSON file
with open('../../../data/books_data_reviews.json', 'r', encoding='utf-8') as file:
    data = json.load(file)
```

```python

# Initialize KeyBERT model
model = KeyBERT()

# Function to extract keywords from a given text
def extract_keywords(text):
    keywords = model.extract_keywords(text)
    return [keyword[0] for keyword in keywords]
```

```python
i = 0
# get time
import time

start_time = time.time()
# Iterate through each book
for book in data:
    print(i)
    i += 1
    # Extract keywords from reviews
    review_keywords = []
    for review in book['reviews']:
        if review['text']:
            review_keywords.extend(extract_keywords(review['text']))

    # Get the top 10 keywords
    top_keywords = review_keywords[:10]

    # Add the 'keywords' field to the book object
    book['keywords'] = top_keywords

    print("Time elapsed: " + str(time.time() - start_time) + ". Estimated time remaining: " + str((time.time() - start_time) / i * (len(data) - i)/60) + " minutes")
```

```python

# Save the modified data back to the JSON file
with open('../../data/books_data_reviews_keywords.json', 'w', encoding='utf-8') as file:
    json.dump(data, file, ensure_ascii=False, indent=4)
```
